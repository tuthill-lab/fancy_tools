```python
# Loading in coordinates from a CSV is of type str so this converts them into a list of integers
pt_ints = []
for i in hps['position_pt'].tolist():
    i = i.split('[')[1]
    i = i.split(']')[0].split(',')
    i = [j for j in i if j != '']
    
    pt_ints.append(list(map(int, i)))
    
hps['position_pt'] = pt_ints
```

```python
# create annotation table
class_sys = hps['Neuromere'] + '_' + hps['Nerve Name']
cell_type = hps['Associated Hair Plate']
pos_pt = hps['position_pt'].values.tolist()
hp_annotation_table = pd.DataFrame(data={'pt_position':pos_pt, 'classification_system': class_sys, 'cell_type': cell_type})
hp_annotation_table
```
